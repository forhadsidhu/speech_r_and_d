```python
import librosa
import librosa.display
import numpy as np
import soundfile as sf
import IPython.display as ipd
import os
```

```python
def download_audio(url,filename):
  """Download an audio file from a given URL"""
  os.system(f"wget -0 {filename}{url}")
```

```python
def load_audio(file_path,sample_rate=16000):
  """Load an audio file and return the waveform and sample rate"""
  audio,sr = librosa.load(file_path,sr=sample_rate)
  return audio,sr
```

```python
def mix_audio(audio1,audio2):
  """Mix two audio signals and normalize the volume"""
  min_len=min(len(audio1),len(audio2))
  mixed = audio1[:min_len]+audio2[:min_len]
  return mixed/np.max(np.abs(mixed)) #Normalized
```

```python
def save_audio(file_path,audio,sample_rate=16000):
  """Save an audio waveform as a WAV file"""
  sf.write(file_path,audio,sample_rate)
```

```python
def play_audio(file_path):
  """Play an audio file in Colab"""
  return ipd.Audio(file_path)
```

```python
from pyannote.audio import Pipeline
import librosa
import soundfile as sf
from pyannote.core import Segment

# Load the diarization model
pipeline = Pipeline.from_pretrained("pyannote/speaker-diarization")

# Run diarization on the audio file
AUDIO_FILE = "/content/overlapping_speakers.wav"
diarization = pipeline(AUDIO_FILE)

# Load the audio using librosa
y, sr = librosa.load(AUDIO_FILE, sr=16000)

# Separate the audio into speaker-specific segments
speaker_segments = {}
for turn, _, speaker in diarization.itertracks(yield_label=True):
    start_time = turn.start
    end_time = turn.end

    # Convert start and end time to sample indices
    start_sample = int(start_time * sr)
    end_sample = int(end_time * sr)

    # Extract the speaker's audio segment
    speaker_audio = y[start_sample:end_sample]

    # Save each speaker's segment to a separate file
    if speaker not in speaker_segments:
        speaker_segments[speaker] = []
    speaker_segments[speaker].append(speaker_audio)

# Save individual speaker's audio files
for speaker, audio_list in speaker_segments.items():
    output_filename = f"speaker_{speaker}_audio.wav"
    # Concatenate audio from the same speaker (if there are multiple segments)
    full_audio = np.concatenate(audio_list, axis=0)
    sf.write(output_filename, full_audio, sr)

    print(f"Saved audio for Speaker {speaker} to {output_filename}")
```
